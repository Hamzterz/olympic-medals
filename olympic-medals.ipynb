# Olympic Medals Analysis

As performed in the coursera course assignment undertaken.

The data about to be imported from olympics.csv was taken from the coursera course which was derived from a wikipedia entry on All Time Olympic Games Medals. Some data cleaning will be done for some practice.

```python
import pandas as pd
```

```python
df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)
```

```python
df.head(8)
```

```python
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2] =='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2] =='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)
```

```python
df.head(8)
```

```python
names_ids = df.index.str.split('\s\(')
names_ids
```

```python
df.index = names_ids.str[0]
```

```python
df.head(8)
```

The columns are organised as # of Summer Games, Summer medals, # of Winter Games, Winter Medals, total # number of games, total # of medals.

```python
df['ID'] = names_ids.str[1].str[:3]
```

```python
df.head(8)
```

```python
df.drop('Totals', inplace=True)
```

### Which country has the biggest difference between their summer and winter gold medal counts?

```python
df['gold_diff'] = abs(df['Gold'] - df['Gold.1'])
```

```python
df['gold_diff'].idxmax()
```

```python
df[df['gold_diff'] == df['gold_diff'].max()]
```

The United States has the largest absolute difference between number of Summer and Winter Gold Medals.

Out of curiosity, which countries do have the largest gap?

```python
df_temp = df.copy()
df_temp.sort_values('gold_diff', ascending=False).head(10)
```

### Which country has the biggest difference between their summer gold medal counts and winter gold medal counts relative to total gold medal count?

```python
df_won_both = df_temp[(df_temp['Gold'] > 0) & (df_temp['Gold.1'] > 0)]
```

```python
df_won_both['relative_difference_gold'] = df_won_both.gold_diff/df_won_both['Gold.2']
```

```python
df_won_both['relative_difference_gold'].idxmax()
```

```python
df_won_both[df_won_both['relative_difference_gold'] == df_won_both['relative_difference_gold'].max()][['Gold', 'Gold.1', 'Gold.2', 'gold_diff', 'relative_difference_gold']]
```

Bulgaria has the largest difference between summer gold and winter gold medal counts, relative to the total gold medal count, by winning 51 golds in Summer Olympics while only winning 1 gold in Winter Olympics (including only counties who have won at least one of both).

Out of curiosity what are the top countries with the largest relative difference in gold medals.

```python
df_won_both.sort_values('relative_difference_gold', ascending=False).head(10)[['Gold', 'Gold.1', 'Gold.2', 'gold_diff', 'relative_difference_gold']]
```

### Tally the points of all country medals by assigning 3 points for gold, 2 for silver, and 1 for bronze.

```python
df['points'] = df['Gold.2']*3 + df['Silver.2']*2 + df['Bronze.2']
```

```python
df['points'].sort_values(ascending=False)[:10]
```

Tally the points of all medals, assigning the points of 3 per gold, 2 per silver, and 1 per bronze, the United States has the most points of any country in the word with 5684 points. This is more than double of the second highest being the Soviet Union, and more than triple of the third higest (Great Britain).
